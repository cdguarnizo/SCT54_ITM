## Ejemplo 1: Un transmisor DSBFC de AM tiene potencia de portadora no modulada $P_𝑐=100𝑊$, que se modula con 3 señales con $𝑚_1=0.2$, $𝑚_2=0.4$, y $𝑚_3=0.5$, determinar  

```
#Cargamos los datos en variables
Pc = 100 #Potencias de la portadora sin modular
m = np.array([0.2,0.4,0.5]) #Vector de indices de modulacion
```

### a) El coeficiente total de modulación. Se calcula usando $m_t = \sqrt{m_1^2+m_2^2+m_3^2}$

```
mt = np.sqrt(np.sum(m**2))
print(mt)
```

### b) La potencia de la bandas laterales superior e inferior. Se calcula usando $P_{tbls}=P_{tbli}=\frac{P_c m_t^2}{2}$

```
Ptbls = Pc*mt**2/2
print(Ptbls)
```

### b) La potencia transmitida. Se calcula usando $P_{t}=P_{c}\left(1+\frac{m_t^2}{2}\right)$

```
Pt = Pc*(1+mt**2/2)
print(Pt)
```

## Ejemplo 2: Encuentre el índice de modulación si la portadora de 10V es modulada por 3 frecuencias diferentes, con amplitudes 1, 2, y 3 voltios, respectivamente.

```
m = Vm/Ec   #Vector de indices de modulacion
mt = np.sqrt(np.sum(m**2))
print(mt)
```

## Ejercicio 1: Una portadora de 1.4MHz es modulada por una señal de música que tiene frecuencias desde 20Hz hasta 10kHz. Determine el rango de frecuencias generadas por las bandas superior e inferior.

```
#Datos
fc = 1.4e6 #Frecuencia de la portadora
fm_min = 20 #Frecuencia minima de la moduladora
fm_max = 10e3 #Frecuencia minima de la moduladora
```

El rango de frecuencias para la banda inferior es: $f_{BLI}(\min) = f_c - f_m(\max)$ y $f_{BLI}(\max) = f_c - f_m(\min)$

```
fbli_min = fc-fm_max
fbli_max = fc-fm_min
print(fbli_min,fbli_max)
```

El rango de frecuencias para la banda inferior es: $f_{BLS}(\min) = f_c + f_m(\min)$ y $f_{BLS}(\max) = f_c - f_m(\max)$

```
fbls_min = fc+fm_min
fbls_max = fc+fm_max
print(fbls_min,fbls_max)
```

## Ejercicio 4: Una emisora AM opera a su máximo permitido de 50kW y a 95% de modulación. Cuanta potencia es transmitida a la señal moduladora? 

```
#Datos
Pt = 50e3
m =.95
```

Para este caso necesitamos calcular primero la potencia de portadora, la cual se calcula despejando $P_c$ de  $P_{t}=P_{c}\left(1+\frac{m^2}{2}\right)$. Entonces, $P_c = P_t/(1+\frac{m^2}{2})$.

```
Pc = Pt/(1+m**2/2)
print(Pc)
```

Entonces, la potencia transmitida a la señal moduladora se calcula: $P_m = P_t - P_c$.

```
Pm = Pt-Pc
print(Pm)
```

## Ejercicio 5: La corriente de una antena de transmisión AM es 12A sin modular pero se incrementa a 13A cuando se modula. Calcular m.

```
#Datos
Ic = 12
It = 13
```

Para este caso tenemos que $I_t = I_c\sqrt{1+\frac{m^2}{2}}$. Despejando $m$ se obtiene, $m = \sqrt{2\left[\left( \frac{I_t}{I_c} \right)^2-1 \right]}$

```
m = np.sqrt(2*((It/Ic)**2 - 1))
print(m)
```
