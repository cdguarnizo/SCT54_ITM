# Modulacion Digital (ASK, FSK, PSK)

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

 

fc = 1e3 # Frecuencia portadora
tb = 10e-3 # Tiempo de bit
deltaf = 300 # delta f
Ac = 1.0 # Amplitud Portadora
fm = fc-deltaf # Frecuencia de marca
fs = fc+deltaf # Frecuencia de espacio

 
v_b = [1,0,1,0,0,1] #Codigo binario

Nbits = len(v_b)
v_tb = np.linspace(0,tb,100)
t = np.linspace(0,Nbits*tb,Nbits*100)
```

## ASK

```
vfm = np.array([])
for k in range(Nbits):
  if v_b[k] is 1:
    vfm = np.append(vfm, Ac*np.cos(2.*np.pi*fc*v_tb))
  else:
    vfm = np.append(vfm, 0.0*np.cos(2.*np.pi*fc*v_tb))
plt.figure(figsize=(20,10))
plt.plot(t,vfm)
```

# FSK

```
vfm = np.array([])
for k in range(Nbits):
  if v_b[k] is 1:
    vfm = np.append(vfm, Ac*np.cos(2.*np.pi*fm*v_tb))
  else:
    vfm = np.append(vfm, Ac*np.cos(2.*np.pi*fs*v_tb))
plt.figure(figsize=(20,10))
plt.plot(t,vfm)
```

## PSK

```
fc = 200
vfm = np.array([])
for k in range(Nbits):
  if v_b[k] is 1:
    vfm = np.append(vfm, Ac*np.cos(2.*np.pi*fc*v_tb))
  else:
    vfm = np.append(vfm, Ac*np.cos(2.*np.pi*fc*v_tb+np.pi))
plt.figure(figsize=(20,10))
plt.plot(t,vfm)
```
